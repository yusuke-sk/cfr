## ライブラリ

```python
import pandas as pd 
import matplotlib.pyplot as plt
```

## データ読み込み

```python
NUM_PLAYERS = 2
```

```python
# -- player2 --
df_1 = pd.read_csv("../output/Leduc_Poker/{}players/DB_for_NFSP_sql_0.1_False.csv".format(NUM_PLAYERS))
df_2 = pd.read_csv("../output/Leduc_Poker/{}players/DB_for_NFSP_sql_1.0_False.csv".format(NUM_PLAYERS))
df_3 = pd.read_csv("../output/Leduc_Poker/{}players/DB_for_NFSP_sql_10.0_False.csv".format(NUM_PLAYERS))
df_4 = pd.read_csv("../output/Leduc_Poker/{}players/DB_for_NFSP_dqn.csv".format(NUM_PLAYERS))

df_4.tail()
```

## 図示

```python
plt.rcParams["font.size"] = 40
plt.figure(figsize=(24,12))
plt.plot(df_1.iteration, df_1.exploitability_for_42_sql, label="NFSP with SQL (alpha = 0.1)" , lw=3, color="r")
plt.plot(df_2.iteration, df_2.exploitability_for_42_sql, label="NFSP with SQL (alpha = 1.0)", lw=3, color="dodgerblue")
plt.plot(df_3.iteration, df_3.exploitability_for_42_sql, label="NFSP with SQL (alpha = 10)" , lw=3, color = "limegreen")
plt.plot(df_4.iteration, df_4.exploitability_for_42_dqn, label="NFSP with DQN", lw=5, color="magenta")

plt.xscale('log')
plt.yscale('log')
plt.xlabel('number of iteration')
plt.ylabel('exploitability')
plt.legend(loc = "lower left")
plt.grid()
plt.xlim(10)
plt.ylim(0.001,10)
plt.savefig("../png/leduc_poker_{}players_nfsp.png".format(NUM_PLAYERS))
```
