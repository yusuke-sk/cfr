## ライブラリ

```python
import pandas as pd 
import matplotlib.pyplot as plt
```

## データ読み込み

```python
NUM_PLAYERS = 3
RANDOM_SEED = [1, 10, 100]
ALPHA = [0.1, 1.0, 10.0, 50.0]
```

## alpha 減衰しないケース

```python
df_a = pd.read_csv("../output/Kuhn_poker/{}players/DB_for_{}_NFSP_sql_{}_False.csv".format(NUM_PLAYERS, RANDOM_SEED[0], ALPHA[0]))
df_b = pd.read_csv("../output/Kuhn_poker/{}players/DB_for_{}_NFSP_sql_{}_False.csv".format(NUM_PLAYERS, RANDOM_SEED[1], ALPHA[0]))
df_c = pd.read_csv("../output/Kuhn_poker/{}players/DB_for_{}_NFSP_sql_{}_False.csv".format(NUM_PLAYERS, RANDOM_SEED[2], ALPHA[0]))
df_1 = df_a.merge(df_b).merge(df_c)
df_1 = df_1.set_index("iteration")
df_1["ex_mean"] = df_1.mean(axis=1)


df_d = pd.read_csv("../output/Kuhn_poker/{}players/DB_for_{}_NFSP_sql_{}_False.csv".format(NUM_PLAYERS, RANDOM_SEED[0], ALPHA[1]))
df_e = pd.read_csv("../output/Kuhn_poker/{}players/DB_for_{}_NFSP_sql_{}_False.csv".format(NUM_PLAYERS, RANDOM_SEED[1], ALPHA[1]))
df_f = pd.read_csv("../output/Kuhn_poker/{}players/DB_for_{}_NFSP_sql_{}_False.csv".format(NUM_PLAYERS, RANDOM_SEED[2], ALPHA[1]))
df_2 = df_d.merge(df_e).merge(df_f)
df_2 = df_2.set_index("iteration")
df_2["ex_mean"] = df_2.mean(axis=1)


df_g = pd.read_csv("../output/Kuhn_poker/{}players/DB_for_{}_NFSP_sql_{}_False.csv".format(NUM_PLAYERS, RANDOM_SEED[0], ALPHA[2]))
df_h = pd.read_csv("../output/Kuhn_poker/{}players/DB_for_{}_NFSP_sql_{}_False.csv".format(NUM_PLAYERS, RANDOM_SEED[1], ALPHA[2]))
df_i = pd.read_csv("../output/Kuhn_poker/{}players/DB_for_{}_NFSP_sql_{}_False.csv".format(NUM_PLAYERS, RANDOM_SEED[2], ALPHA[2]))
df_3 = df_g.merge(df_h).merge(df_i)
df_3 = df_3.set_index("iteration")
df_3["ex_mean"] = df_3.mean(axis=1)


df_j = pd.read_csv("../output/Kuhn_poker/{}players/DB_for_{}_NFSP_dqn.csv".format(NUM_PLAYERS, RANDOM_SEED[0]))
df_k = pd.read_csv("../output/Kuhn_poker/{}players/DB_for_{}_NFSP_dqn.csv".format(NUM_PLAYERS, RANDOM_SEED[1]))
df_l = pd.read_csv("../output/Kuhn_poker/{}players/DB_for_{}_NFSP_dqn.csv".format(NUM_PLAYERS, RANDOM_SEED[2]))
df_4 = df_j.merge(df_k).merge(df_l)
df_4 = df_4.set_index("iteration")
df_4["ex_mean"] = df_4.mean(axis=1)



plt.rcParams["font.size"] = 40
plt.figure(figsize=(24,15))
plt.plot(df_1.index, df_1.ex_mean, label="NFSP with SQL ({} = {})".format(chr(945), ALPHA[0]) , lw=6, color="k")
plt.plot(df_2.index, df_2.ex_mean, label="NFSP with SQL ({} = {})".format(chr(945),ALPHA[1]) , lw=6, color="b")
plt.plot(df_3.index, df_3.ex_mean, label="NFSP with SQL ({} = {})".format(chr(945),ALPHA[2]) , lw=6, color = "g")
plt.plot(df_4.index, df_4.ex_mean, label="NFSP with DQN", lw=6, color="r")


plt.xscale('log')
plt.yscale('log')
plt.xlabel('number of iterations', labelpad=20)
plt.ylabel('exploitability', labelpad=20)
plt.legend(loc = "lower left", facecolor="white", edgecolor="white", framealpha=1)
plt.grid(linestyle = "--", linewidth = 1.5)
plt.gca().spines["top"].set_linewidth(1.5)
plt.gca().spines["left"].set_linewidth(1.5)
plt.gca().spines["bottom"].set_linewidth(1.5)
plt.gca().spines["right"].set_linewidth(1.5)
plt.xlim(10)
plt.tick_params(pad=20)
plt.ylim(0.01,10)
plt.tight_layout()
plt.tick_params(width = 1, length = 15, which = "minor", direction = "in")

plt.savefig("../png/kuhn_poker_{}players_SQL_alpha_not_discrease.png".format(NUM_PLAYERS))
```

## alpha 減衰するケース

```python
df_a = pd.read_csv("../output/Kuhn_poker/{}players/DB_for_{}_NFSP_sql_{}_True.csv".format(NUM_PLAYERS, RANDOM_SEED[0], ALPHA[2]))
df_b = pd.read_csv("../output/Kuhn_poker/{}players/DB_for_{}_NFSP_sql_{}_True.csv".format(NUM_PLAYERS, RANDOM_SEED[1], ALPHA[2]))
df_c = pd.read_csv("../output/Kuhn_poker/{}players/DB_for_{}_NFSP_sql_{}_True.csv".format(NUM_PLAYERS, RANDOM_SEED[2], ALPHA[2]))
df_1 = df_a.merge(df_b).merge(df_c)
df_1 = df_1.set_index("iteration")
df_1["ex_mean"] = df_1.mean(axis=1)


df_d = pd.read_csv("../output/Kuhn_poker/{}players/DB_for_{}_NFSP_sql_{}_True.csv".format(NUM_PLAYERS, RANDOM_SEED[0], ALPHA[3]))
df_e = pd.read_csv("../output/Kuhn_poker/{}players/DB_for_{}_NFSP_sql_{}_True.csv".format(NUM_PLAYERS, RANDOM_SEED[1], ALPHA[3]))
df_f = pd.read_csv("../output/Kuhn_poker/{}players/DB_for_{}_NFSP_sql_{}_True.csv".format(NUM_PLAYERS, RANDOM_SEED[2], ALPHA[3]))
df_2 = df_d.merge(df_e).merge(df_f)
df_2 = df_2.set_index("iteration")
df_2["ex_mean"] = df_2.mean(axis=1)


df_g = pd.read_csv("../output/Kuhn_poker/{}players/DB_for_{}_NFSP_dqn.csv".format(NUM_PLAYERS, RANDOM_SEED[0]))
df_h = pd.read_csv("../output/Kuhn_poker/{}players/DB_for_{}_NFSP_dqn.csv".format(NUM_PLAYERS, RANDOM_SEED[1]))
df_i = pd.read_csv("../output/Kuhn_poker/{}players/DB_for_{}_NFSP_dqn.csv".format(NUM_PLAYERS, RANDOM_SEED[2]))
df_3 = df_g.merge(df_h).merge(df_i)
df_3 = df_3.set_index("iteration")
df_3["ex_mean"] = df_3.mean(axis=1)



plt.rcParams["font.size"] = 40
plt.figure(figsize=(24,15))
plt.plot(df_1.index, df_1.ex_mean, label="NFSP with SQL ({}_initial = {})".format(chr(945), ALPHA[2]) , lw=6, color="b")
plt.plot(df_2.index, df_2.ex_mean, label="NFSP with SQL ({}_initial = {})".format(chr(945), ALPHA[3]) , lw=6, color="g")
plt.plot(df_3.index, df_3.ex_mean, label="NFSP with DQN", lw=6, color="r")

plt.xscale('log')
plt.yscale('log')
plt.xlabel('number of iterations', labelpad=20)
plt.ylabel('exploitability', labelpad=20)
plt.legend(loc = "lower left", facecolor="white", edgecolor="white", framealpha=1)
plt.grid(linestyle = "--", linewidth = 1.5)
plt.gca().spines["top"].set_linewidth(1.5)
plt.gca().spines["left"].set_linewidth(1.5)
plt.gca().spines["bottom"].set_linewidth(1.5)
plt.gca().spines["right"].set_linewidth(1.5)
plt.xlim(10)
plt.tick_params(pad=20)
plt.ylim(0.01,10)
plt.tight_layout()
plt.tick_params(width = 1, length = 15, which = "minor", direction = "in")

plt.savefig("../png/kuhn_poker_{}players_SQL_alpha_discrease.png".format(NUM_PLAYERS))
```

## 他手法と比較

```python
df_a = pd.read_csv("../output/Kuhn_poker/{}players/DB_for_{}_chance_sampling_CFR.csv".format(NUM_PLAYERS, RANDOM_SEED[0]))
df_b = pd.read_csv("../output/Kuhn_poker/{}players/DB_for_{}_chance_sampling_CFR.csv".format(NUM_PLAYERS, RANDOM_SEED[1]))
df_c = pd.read_csv("../output/Kuhn_poker/{}players/DB_for_{}_chance_sampling_CFR.csv".format(NUM_PLAYERS, RANDOM_SEED[2]))
df_1 = df_a.merge(df_b).merge(df_c)
df_1 = df_1.set_index("iteration")
df_1["ex_mean"] = df_1.mean(axis=1)

df_d = pd.read_csv("../output/Kuhn_poker/{}players/DB_for_{}_outcome_sampling_MCCFR.csv".format(NUM_PLAYERS, RANDOM_SEED[0]))
df_e = pd.read_csv("../output/Kuhn_poker/{}players/DB_for_{}_outcome_sampling_MCCFR.csv".format(NUM_PLAYERS, RANDOM_SEED[1]))
df_f = pd.read_csv("../output/Kuhn_poker/{}players/DB_for_{}_outcome_sampling_MCCFR.csv".format(NUM_PLAYERS, RANDOM_SEED[2]))
df_2 = df_d.merge(df_e).merge(df_f)
df_2 = df_2.set_index("iteration")
df_2["ex_mean"] = df_2.mean(axis=1)

df_g = pd.read_csv("../output/Kuhn_poker/{}players/DB_for_{}_FSP.csv".format(NUM_PLAYERS, RANDOM_SEED[0]))
df_h = pd.read_csv("../output/Kuhn_poker/{}players/DB_for_{}_FSP.csv".format(NUM_PLAYERS, RANDOM_SEED[1]))
df_i = pd.read_csv("../output/Kuhn_poker/{}players/DB_for_{}_FSP.csv".format(NUM_PLAYERS, RANDOM_SEED[2]))
df_3 = df_g.merge(df_h).merge(df_i)
df_3 = df_3.set_index("iteration")

df_3 = df_g.set_index("iteration")
df_3["ex_mean"] = df_3.mean(axis=1)




df_j = pd.read_csv("../output/Kuhn_poker/{}players/DB_for_{}_NFSP_sql_{}_True.csv".format(NUM_PLAYERS, RANDOM_SEED[0], ALPHA[3]))
df_k = pd.read_csv("../output/Kuhn_poker/{}players/DB_for_{}_NFSP_sql_{}_True.csv".format(NUM_PLAYERS, RANDOM_SEED[1], ALPHA[3]))
df_l = pd.read_csv("../output/Kuhn_poker/{}players/DB_for_{}_NFSP_sql_{}_True.csv".format(NUM_PLAYERS, RANDOM_SEED[2], ALPHA[3]))
df_4 = df_j.merge(df_k).merge(df_l)
df_4 = df_4.set_index("iteration")
df_4["ex_mean"] = df_4.mean(axis=1)


df_m = pd.read_csv("../output/Kuhn_poker/{}players/DB_for_{}_NFSP_dqn.csv".format(NUM_PLAYERS, RANDOM_SEED[0]))
df_n = pd.read_csv("../output/Kuhn_poker/{}players/DB_for_{}_NFSP_dqn.csv".format(NUM_PLAYERS, RANDOM_SEED[1]))
df_o = pd.read_csv("../output/Kuhn_poker/{}players/DB_for_{}_NFSP_dqn.csv".format(NUM_PLAYERS, RANDOM_SEED[2]))
df_5 = df_m.merge(df_n).merge(df_o)
df_5 = df_5.set_index("iteration")
df_5["ex_mean"] = df_5.mean(axis=1)


plt.rcParams["font.size"] = 35
plt.figure(figsize=(24,15))
plt.plot(df_1.index, df_1.ex_mean, label="Chance Sampling CFR" , lw=6, color="g")
plt.plot(df_2.index, df_2.ex_mean, label="Outcome Sampling MCCFR", lw=6, color="r")
plt.plot(df_3.index, df_3.ex_mean, label="FSP", lw=6, color="b")
plt.plot(df_4.index, df_4.ex_mean, label="NFSP with SQL ({}_initial = {})".format(chr(945), ALPHA[3]), lw=6, color="m")
plt.plot(df_5.index, df_5.ex_mean, label="NFSP with DQN", lw=6, color="k")


plt.xscale('log')
plt.yscale('log')
plt.xlabel('number of iterations', labelpad=20)
plt.ylabel('exploitability', labelpad=20)
plt.legend(loc = "lower left", facecolor="white", edgecolor="white", framealpha=1)
plt.grid(linestyle = "--", linewidth = 1.5)
plt.gca().spines["top"].set_linewidth(1.5)
plt.gca().spines["left"].set_linewidth(1.5)
plt.gca().spines["bottom"].set_linewidth(1.5)
plt.gca().spines["right"].set_linewidth(1.5)
plt.xlim(10)
plt.tick_params(pad=20)
plt.ylim(0.001,10)
plt.tight_layout()
plt.tick_params(width = 1, length = 15, which = "minor", direction = "in")
plt.savefig("../png/kuhn_poker_{}players_all_method.png".format(NUM_PLAYERS))
```
