## ライブラリ

```python
import pandas as pd 
import matplotlib.pyplot as plt
```

## データ読み込み

```python
# -- player2 --
df_1 = pd.read_csv("../output/2players/DB_for_NFSP_sql_0.1_False.csv")
df_2 = pd.read_csv("../output/2players/DB_for_NFSP_sql_1.0_False.csv")
df_3 = pd.read_csv("../output/2players/DB_for_NFSP_sql_1e-12_False.csv")
df_4 = pd.read_csv("../output/2players/DB_for_NFSP_dqn.csv")


df = pd.concat([df_1, df_2, df_3, df_4], axis=1, join='outer')

df = df.loc[:,~df.columns.duplicated()]
df = df.set_index('iteration')


df.tail()
```

## 図示

```python
plt.rcParams["font.size"] = 40
plt.figure(figsize=(24,12))
plt.plot(df.index, df.exploitability_rate_2_mean, label="2-Kuhn exploit" , lw=5, color="r")
plt.plot(df.index, df.exploitability_rate_3_mean, label="3-Kuhn exploit", lw=5 color="dodgerblue")
plt.plot(df.index, df.exploitability_rate_4_mean, label="4-Kuhn exploit" , lw=5, color = "limegreen")
plt.plot(df.index, df.exploitability_rate_5_mean, label="5-Kuhn exploit", lw=5, color="magenta")

plt.xscale('log')
plt.yscale('log')
plt.xlabel('iteration')
plt.ylabel('exploitability rate')
plt.legend(loc = "lower left")
plt.grid()
plt.xlim(10)
plt.ylim(0.001,10)
plt.savefig("./png/kuhn_poker_nfsp.png")
```
