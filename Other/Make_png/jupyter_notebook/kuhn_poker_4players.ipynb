## ライブラリ

```python
import pandas as pd 
import matplotlib.pyplot as plt
```

## データ読み込み

```python
NUM_PLAYERS = 4
```

## alpha 減衰しないケース

```python
# -- player4 --
df_1 = pd.read_csv("../output/Kuhn_poker/{}players/DB_for_NFSP_sql_0.1_False.csv".format(NUM_PLAYERS))
df_2 = pd.read_csv("../output/Kuhn_poker/{}players/DB_for_NFSP_sql_1.0_False.csv".format(NUM_PLAYERS))
df_3 = pd.read_csv("../output/Kuhn_poker/{}players/DB_for_NFSP_sql_10.0_False.csv".format(NUM_PLAYERS))
df_4 = pd.read_csv("../output/Kuhn_poker/{}players/DB_for_NFSP_dqn.csv".format(NUM_PLAYERS))


df_3.tail()

plt.rcParams["font.size"] = 40
plt.figure(figsize=(24,15))
plt.plot(df_1.iteration, df_1.exploitability_for_42_sql, label="NFSP with SQL (alpha = 0.1)" , lw=2, color="k")
plt.plot(df_2.iteration, df_2.exploitability_for_42_sql, label="NFSP with SQL (alpha = 1.0)", lw=2, color="b")
plt.plot(df_3.iteration, df_3.exploitability_for_42_sql, label="NFSP with SQL (alpha = 10.0)" , lw=2, color = "g")
plt.plot(df_4.iteration, df_4.exploitability_for_42_dqn, label="NFSP with DQN", lw=2, color="r")


plt.xscale('log')
plt.yscale('log')
plt.xlabel('number of iterations', labelpad=20)
plt.ylabel('exploitability', labelpad=20)
plt.legend(loc = "lower left", facecolor="white", edgecolor="white", framealpha=1)
plt.grid(linestyle = "--", linewidth = 1.5)
plt.gca().spines["top"].set_linewidth(1.5)
plt.gca().spines["left"].set_linewidth(1.5)
plt.gca().spines["bottom"].set_linewidth(1.5)
plt.gca().spines["right"].set_linewidth(1.5)
plt.xlim(10)
plt.tick_params(pad=20)
plt.ylim(0.01,10)
plt.tight_layout()
plt.tick_params(width = 1, length = 15, which = "minor", direction = "in")

plt.savefig("../png/kuhn_poker_{}players_SQL_alpha_not_discrease.png".format(NUM_PLAYERS))
```

## alpha 減衰するケース

```python
df_1 = pd.read_csv("../output/Kuhn_poker/{}players/DB_for_NFSP_sql_10.0_True.csv".format(NUM_PLAYERS))
df_2 = pd.read_csv("../output/Kuhn_poker/{}players/DB_for_NFSP_sql_50.0_True.csv".format(NUM_PLAYERS))
df_3 = pd.read_csv("../output/Kuhn_poker/{}players/DB_for_NFSP_dqn.csv".format(NUM_PLAYERS))

df_3.tail()

plt.rcParams["font.size"] = 40
plt.figure(figsize=(24,15))
plt.plot(df_1.iteration, df_1.exploitability_for_42_sql, label="NFSP with SQL (initial_alpha = 10.0)" , lw=2, color="b")
plt.plot(df_2.iteration, df_2.exploitability_for_42_sql, label="NFSP with SQL (initial_alpha = 50.0)", lw=2, color="g")
plt.plot(df_3.iteration, df_4.exploitability_for_42_dqn, label="NFSP with DQN", lw=2, color="r")

plt.xscale('log')
plt.yscale('log')
plt.xlabel('number of iterations', labelpad=20)
plt.ylabel('exploitability', labelpad=20)
plt.legend(loc = "lower left", facecolor="white", edgecolor="white", framealpha=1)
plt.grid(linestyle = "--", linewidth = 1.5)
plt.gca().spines["top"].set_linewidth(1.5)
plt.gca().spines["left"].set_linewidth(1.5)
plt.gca().spines["bottom"].set_linewidth(1.5)
plt.gca().spines["right"].set_linewidth(1.5)
plt.xlim(10)
plt.tick_params(pad=20)
plt.ylim(0.01,10)
plt.tight_layout()
plt.tick_params(width = 1, length = 15, which = "minor", direction = "in")
plt.savefig("../png/kuhn_poker_{}players_SQL_alpha_discrease.png".format(NUM_PLAYERS))
```

## 他手法と比較

```python
df_1 = pd.read_csv("../output/Kuhn_poker/{}players/DB_for_42_chance_sampling_CFR.csv".format(NUM_PLAYERS))
df_2 = pd.read_csv("../output/Kuhn_poker/{}players/DB_for_42_outcome_sampling_MCCFR.csv".format(NUM_PLAYERS))
df_3 = pd.read_csv("../output/Kuhn_poker/{}players/DB_for_42_FSP.csv".format(NUM_PLAYERS))
df_4 = pd.read_csv("../output/Kuhn_poker/{}players/DB_for_NFSP_sql_50.0_True.csv".format(NUM_PLAYERS))
df_5 = pd.read_csv("../output/Kuhn_poker/{}players/DB_for_NFSP_dqn.csv".format(NUM_PLAYERS))

#display(df_3)

plt.rcParams["font.size"] = 35
plt.figure(figsize=(24,15))
plt.plot(df_1.iteration, df_1.exploitability_for_42_chance_sampling_CFR, label="Chance Sampling CFR" , lw=2, color="green")
plt.plot(df_2.iteration, df_2.exploitability_for_42_outcome_sampling_MCCFR, label="Outcome Sampling MCCFR", lw=2, color="purple")
plt.plot(df_3.iteration, df_3.exploitability_for_42_FSP, label="FSP", lw=2, color="black")
plt.plot(df_4.iteration, df_4.exploitability_for_42_sql, label="NFSP with SQL (initial_alpha = 50.0)", lw=2, color="dodgerblue")
plt.plot(df_5.iteration, df_5.exploitability_for_42_dqn, label="NFSP with DQN", lw=2, color="magenta")

plt.xscale('log')
plt.yscale('log')
plt.xlabel('number of iterations', labelpad=20)
plt.ylabel('exploitability', labelpad=20)
plt.legend(loc = "lower left", facecolor="white", edgecolor="white", framealpha=1)
plt.grid(linestyle = "--", linewidth = 1.5)
plt.gca().spines["top"].set_linewidth(1.5)
plt.gca().spines["left"].set_linewidth(1.5)
plt.gca().spines["bottom"].set_linewidth(1.5)
plt.gca().spines["right"].set_linewidth(1.5)
plt.xlim(10)
plt.tick_params(pad=20)
plt.ylim(0.001,10)
plt.tight_layout()
plt.tick_params(width = 1, length = 15, which = "minor", direction = "in")
plt.savefig("../png/kuhn_poker_{}players_all_method.png".format(NUM_PLAYERS))
```
